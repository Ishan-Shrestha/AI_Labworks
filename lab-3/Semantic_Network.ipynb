## Theory:

### Natural Language Processing (NLP) Pipeline

To extract recommendation features from unstructured academic text (such as titles, abstracts, or metadata) and convert them into a structured engine, several sequential NLP tasks are performed:

* **Part-of-Speech (POS) Tagging:** The NLP model parses text tokens and flags grammatical components (Proper Nouns, Verbs, Adjectives). In a recommendation context, this isolates core domain actions (e.g., "utilizes", "improves") from filler words.
* **Named Entity Recognition (NER):** This identifies and classifies real-world entities into predefined categories. In an academic system, it extracts specialized entities like **Authors**, **Methodologies**, and **Research Domains**.
* **Vector Embedding (Word2Vec):** High-dimensional textual tokens are transformed into a dense, low-dimensional vector space. By mapping semantic meaning numerically, it allows the system to calculate the contextual similarity between different research concepts even if they do not share exact keyword matches.

### Knowledge Triples as Recommendation Vectors

Knowledge triples serve as the fundamental relational building blocks of the system. They convert raw NLP attributes into explicit, machine-readable facts by combining tokens, structural syntax (POS), and categorical classifications (NER):

```text
       predicate
Subject ───────► Object

```

* **Subject (Source Node):** The entity initiating the relationship (e.g., a specific `Paper` or `Author`).
* **Predicate (Edge Label):** The directed relationship or action linking the entities (e.g., `CITES`, `UTILIZED`, `AUTHORED`).
* **Object (Target Node):** The receiving entity completing the semantic statement (e.g., a `Methodology` or another `Paper`).

### Heterogeneous Semantic Networks (Knowledge Graphs)

Unlike homogeneous graphs where all nodes and edges represent the same entity type (e.g., a pure social network of users), a recommendation system built on academic literature requires a **Heterogeneous Information Network (HIN)**:

* **Nodes (Multi-type):** Represent distinct entity classes within the ecosystem (`Paper`, `Author`, `Methodology`, `Domain`).
* **Edges (Multi-relational):** Represent directed, semantic links mapping how these distinct entities interact.

By organizing data by relationship rather than flat structural storage, the network mimics human memory topology. This allows the recommendation engine to perform graph traversal paths—such as finding papers written by the same author that utilize the exact same methodology.

### Graph-Based Similarity Metrics for Recommendation

To generate recommendations, the system relies on calculating proximity and overlap between nodes using two distinct mathematical frameworks:

#### A. Cosine Similarity (Semantic Vector Proximity)

Used primarily to evaluate content-based similarity via text embeddings. It measures the cosine of the angle between two multi-dimensional vectors, evaluating directional alignment rather than magnitude:

$$\text{Cosine Similarity}(A, B) = \frac{A \cdot B}{\|A\| \|B\|}$$

In recommendations, a score approaching $1$ indicates that two concepts or papers share a tight contextual and semantic meaning within the vector space.

#### B. Jaccard Similarity (Topological Neighborhood Overlap)

Used to evaluate structural similarity based on graph connectivity. It measures the diversity of shared sample sets (e.g., the intersection of shared authors, methodologies, or citations) divided by the total unique pool of attributes between two nodes:

$$\text{Jaccard Similarity}(A, B) = \frac{|A \cap B|}{|A \cup B|}$$

In recommendations, if Paper A and Paper B have a high Jaccard score, it implies they share a significant statistical overlap in their underlying metadata footprint (e.g., they cite the same papers and utilize the same algorithms), making them prime candidates to recommend to a user.

## Program Used:

````python
import random
import json

print(" Step 1: Synthesizing interconnected 30-Paper Research Network...")

# Define structural metadata pools
domains = {
    "CV": ("Computer Vision", "Visual intelligence and spatial feature processing."),
    "NLP": ("Natural Language Processing", "Computational linguistics and sequential text parsing."),
    "Graph_AI": ("Geometric Deep Learning", "Graph-structured data processing and non-Euclidean analytics.")
}

methodologies = {
    "Transformers": ("Transformer Architectures", "Self-attention mechanisms optimized for parallel sequence modeling."),
    "CNNs": ("Convolutional Neural Networks", "Mathematical weight-sharing kernels optimized for spatial grid arrays."),
    "GNNs": ("Graph Neural Networks", "Message-passing node-state configurations optimized for relational topologies."),
    "CL": ("Contrastive Learning", "Self-supervised vector alignment maximizing mutual information across positive pairs.")
}

authors = ["Dr_Y_Bengio", "Dr_G_Hinton", "Dr_Y_LeCun", "Dr_A_Vaswani", "Dr_J_Leskovec"]

# Generate 30 contextual research paper titles distributed across domains
paper_titles = [
    # Computer Vision Papers (1-10)
    ("ResNet-3D: Residual Grids for Spatial Volumetric Analysis", "CV", "CNNs", "Dr_Y_LeCun"),
    ("ViT-Scale: Scaling Vision Transformers for Dense Prediction", "CV", "Transformers", "Dr_A_Vaswani"),
    ("SimCLR-V3: Contrastive Frameworks for Self-Supervised Vision", "CV", "CL", "Dr_G_Hinton"),
    ("Masked Autoencoders Are Scalable Vision Visualizers", "CV", "Transformers", "Dr_Y_LeCun"),
    ("NeRF-Graph: Neural Radiance Fields on Graph Topologies", "CV", "GNNs", "Dr_J_Leskovec"),
    ("Real-Time Semantic Segmentation via Dilated Spatial Kernels", "CV", "CNNs", "Dr_Y_LeCun"),
    ("Cross-Modal Vector Alignment in Generative Vision Spaces", "CV", "CL", "Dr_G_Hinton"),
    ("Attention-Driven Object Detection in High-Resolution Imagery", "CV", "Transformers", "Dr_A_Vaswani"),
    ("Unsupervised Visual Representation via Contrastive Residuals", "CV", "CL", "Dr_G_Hinton"),
    ("Geometry-Aware Convolutions for 3D Point Cloud Processing", "CV", "CNNs", "Dr_J_Leskovec"),
    
    # NLP Papers (11-20)
    ("BERT-Large: Pre-training of Deep Bidirectional Transformers", "NLP", "Transformers", "Dr_A_Vaswani"),
    ("GPT-Next: Autoregressive Language Modeling at Scale", "NLP", "Transformers", "Dr_A_Vaswani"),
    ("Contrastive Sentence Embeddings via Semantic Invariance", "NLP", "CL", "Dr_Y_Bengio"),
    ("Long-Short Sequence Parsing via Linear Attention Windows", "NLP", "Transformers", "Dr_A_Vaswani"),
    ("Text-Graph Recurrent Transformers for Structured Document Analysis", "NLP", "GNNs", "Dr_J_Leskovec"),
    ("Exploring the Limits of Self-Supervised Machine Translation", "NLP", "CL", "Dr_Y_Bengio"),
    ("Dependency-Parsing Convolutions for Low-Resource Languages", "NLP", "CNNs", "Dr_Y_LeCun"),
    ("Token-Free Language Representations via Character-Level Processing", "NLP", "Transformers", "Dr_Y_Bengio"),
    ("Retrieval-Augmented Transformers for Dynamic Knowledge Graphs", "NLP", "Transformers", "Dr_J_Leskovec"),
    ("Evaluating Prompt Calibration Metrics in LLM Quantization", "NLP", "Transformers", "Dr_Y_Bengio"),

    # Graph AI Papers (21-30)
    ("GCN-V2: Scalable Graph Convolutional Networks via Node Sampling", "Graph_AI", "GNNs", "Dr_J_Leskovec"),
    ("Graph Attention Networks with Multi-Head Structural Alignment", "Graph_AI", "Transformers", "Dr_A_Vaswani"),
    ("Self-Supervised Graph Contrastive Learning via Subgraph Masking", "Graph_AI", "CL", "Dr_Y_Bengio"),
    ("Message-Passing Frameworks for Heterogeneous Molecular Topologies", "Graph_AI", "GNNs", "Dr_G_Hinton"),
    ("Graph Convolutions Meet Vision Transformers: A Unified Review", "Graph_AI", "Transformers", "Dr_Y_LeCun"),
    ("Temporal Graph Networks for Dynamic Relational Interaction Stream", "Graph_AI", "GNNs", "Dr_J_Leskovec"),
    ("Inductive Representation Learning on Large-Scale Social Webs", "Graph_AI", "GNNs", "Dr_J_Leskovec"),
    ("Contrastive Node Clustering via Graph Diffusion Wavelets", "Graph_AI", "CL", "Dr_G_Hinton"),
    ("Deep Generative Models for Graph Structure Deconvolution", "Graph_AI", "CNNs", "Dr_Y_LeCun"),
    ("Scalable Graph Neural Architecture Search via Gradient Descent", "Graph_AI", "GNNs", "Dr_Y_Bengio")
]

nodes_dict = {}
edges_set = set()

# Process Domain Metadata Nodes
for dom_key, (dom_name, dom_desc) in domains.items():
    nodes_dict[dom_name] = {
        "group": "Domain", 
        "category": "Academic Field", 
        "rec": "Review literature cross-sections to find multi-disciplinary opportunities."
    }

# Process Methodology Metadata Nodes
for meth_key, (meth_name, meth_desc) in methodologies.items():
    nodes_dict[meth_name] = {
        "group": "Methodology", 
        "category": "Core Algorithm", 
        "rec": f"Lab deployment vector: Isolate this framework to benchmark operational runtime bottlenecks."
    }

# Process Author Metadata Nodes
for author in authors:
    nodes_dict[author] = {
        "group": "Author", 
        "category": "Principal Investigator", 
        "rec": "Analyze h-index and structural degree co-authorship paths to optimize grant sourcing."
    }

# Process the 30 Research Papers into Nodes and Map Structural Affiliation Edges
for idx, (title, dom_key, meth_key, author) in enumerate(paper_titles):
    dom_name = domains[dom_key][0]
    meth_name = methodologies[meth_key][0]
    
    nodes_dict[title] = {
        "group": "Paper",
        "category": f"Published under {dom_name}",
        "rec": f"Literature Mapping Blueprint: Verify citation weights to evaluate the foundational novelty of this work."
    }
    
    # Establish Metadata Semantic Edges
    edges_set.add((title, dom_name, "BELONGS_TO"))
    edges_set.add((title, meth_name, "UTILIZED"))
    edges_set.add((author, title, "AUTHORED"))

# Generate Dense Multi-Layer Internal Citations (Ensures Interconnectedness)
# Lower indices represent older foundational papers, higher indices represent newer works
for i in range(5, len(paper_titles)):
    current_title = paper_titles[i][0]
    current_dom = paper_titles[i][1]
    
    # Target older papers to simulate real retrospective academic referencing
    possible_citations = paper_titles[:i]
    
    # Strategy: High probability to cite papers within the same domain, lower probability to cross domains
    same_domain_pool = [p[0] for p in possible_citations if p[1] == current_dom]
    cross_domain_pool = [p[0] for p in possible_citations if p[1] != current_dom]
    
    citations_to_add = []
    if same_domain_pool:
        citations_to_add.extend(random.sample(same_domain_pool, k=min(len(same_domain_pool), random.randint(1, 3))))
    if cross_domain_pool and random.random() > 0.6:  # 40% chance of cross-pollination
        citations_to_add.append(random.choice(cross_domain_pool))
        
    for cited_paper in citations_to_add:
        if current_title != cited_paper:
            edges_set.add((current_title, cited_paper, "CITES"))

# Format JSON structural layout for WebGL engine
json_data = {
    "nodes": [{"id": name, "group": meta["group"], "category": meta["category"], "rec": meta["rec"]} for name, meta in nodes_dict.items()],
    "links": [{"source": src, "target": tgt, "label": rel} for src, tgt, rel in edges_set]
}

with open("network_data.json", "w") as f:
    json.dump(json_data, f, indent=4)

print(f"✅ Literature Citation Network mapped successfully! Calculated {len(nodes_dict)} nodes and {len(edges_set)}structural paths.")


def get_features(paper_title, network_json):
    """Extracts all connected metadata/citations for a given paper to use as features."""
    features = set()
    for link in network_json["links"]:
        # If the paper is the source (e.g., Paper CITES X, Paper UTILIZED Y)
        if link["source"] == paper_title:
            features.add(link["target"])
        # If the paper is the target (e.g., Author AUTHORED Paper)
        elif link["target"] == paper_title:
            features.add(link["source"])
    return features

def recommend_papers(target_paper, network_json, top_n=3):
    """Recommends top N papers based on Jaccard Similarity of features."""
    # Verify the target paper exists
    all_papers = [n["id"] for n in network_json["nodes"] if n["group"] == "Paper"]
    if target_paper not in all_papers:
        return f"Paper '{target_paper}' not found in the network."
    
    target_features = get_features(target_paper, network_json)
    if not target_features:
        return "No features found for this paper."

    scores = []
    for paper in all_papers:
        if paper == target_paper:
            continue
        
        paper_features = get_features(paper, network_json)
        
        # Calculate Jaccard Similarity
        intersection = len(target_features.intersection(paper_features))
        union = len(target_features.union(paper_features))
        
        jaccard_score = intersection / union if union > 0 else 0
        scores.append((paper, jaccard_score))
    
    # Sort by score descending
    scores.sort(key=lambda x: x[1], reverse=True)
    return scores[:top_n]

# --- Test the Recommendation System ---
print("\n--- Testing Recommendation System ---")
test_paper = str(input("Enter the paper:"))
recommendations = recommend_papers(test_paper, json_data, top_n=3)

print(f"\nIf you liked '{test_paper}', you might also like:")
for rank, (paper, score) in enumerate(recommendations, 1):
    print(f"{rank}. {paper} (Similarity Score: {score:.3f})")
````

## Discussion

In this lab, we successfully transitioned from foundational NLP concept notes to a fully functional Knowledge Graph-Based Recommendation System. The implementation highlights how structured relationships can overcome the limitations of traditional keyword-based search engines.

De-confusing the Link Topology: A critical observation during development was understanding node directionality within a heterogeneous information network. Because relationships are directional, our feature extraction algorithm (get_features) had to dynamically evaluate both the source and target properties of each link. Failing to account for this bi-directional relationship would omit crucial metadata (like the author's identity), severely crippling the recommendation quality.

The Power of Jaccard Overlap: By gathering all incoming and outgoing connections into a unified feature set, we applied Jaccard Similarity to calculate structural proximity. When a user queries a paper like GPT-Next, the system evaluates the mathematical intersection and union of its attributes against the other 29 papers. Papers sharing identical methodologies (e.g., Transformers), domains (e.g., NLP), or authors (e.g., Dr_A_Vaswani) naturally yield higher similarity coefficients (approaching 1.0), making them highly accurate recommendations.
